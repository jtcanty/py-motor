```python
import numpy as np
import pystep as stp
import matplotlib.pyplot as plt
import mpld3
mpld3.enable_notebook()
```

```python
sim = stp.Simulation()
```

```python
trace = sim.build()
```

```python
fit = sim.fit(trace[1])
```

```python
# Plot selected trace and fit
plt.plot(trace[1])
plt.plot(fit[0])
plt.xlabel('Frame')
plt.ylabel('Displacement (nm)')
plt.grid(True, axis='y', linestyle='--')
plt.show()
```
